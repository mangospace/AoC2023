```python
import regex as re
games=[]
with open('day2.txt', 'r') as f:
    for line in f:
        print(line, end='')
        line=re.sub("Game \d+: ","",line) 
        atom=line.split(";")
        print(atom)
        blue=[]
        red=[]
        green=[]
        for x in atom:
            if re.search("(\d+) blue", x):
                blue.append(int(re.search("(\d+) blue", x).group(1)))
            if re.search("(\d+) green", x):
                green.append(int(re.search("(\d+) green", x).group(1)))
            if re.search("(\d+) red", x):
                red.append(int(re.search("(\d+) red", x).group(1)))
        games.append([max(red),  max(green) , max(blue)])

print(games)
```

```python
maxr= 12
maxg=13
maxb=14

possib=0
for index, x in enumerate(games):
    index1=index+1
    if int(x[0]) <= maxr:
        if int(x[1]) <= maxg:
            if int(x[2]) <= maxb:
                possib+=index1
                print(f"game possible: {index+1}")

print(possib)
    

```
