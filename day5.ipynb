```python
import regex as re
p = re.compile("[$*+#]")

s2smap=[[50, 98, 2],[52 ,50, 48]]


with open('day5_short_test.txt', 'r') as f:
    for index, line in enumerate(f):
        linedict[index]=re.sub("\\n","",line)
        rowlen+=1
        lenline=len(line)
        print(f"{line=}")
        symbolist=[]
        if re.findall("[$*+#]",line):
            for m in p.finditer(line):
                posx1= m.start()
                posx2= m.start()
                try:
                    posx1=posx1-1
                    if posx1<0:
                        posx1=0
                except:
                    pass
                try:
                    posx2=posx2+2
                    if posx2>lenline:
                        posx2=lenline
                except:
                    pass
                symbolxset=[x for x in range(posx1, posx2)]
#                print(f"{posx1=}, {posx2=}, {symbolxset=}")
                symbolist.append(symbolxset)
#                print(f"{symbolist=}")
                
        numlist=[]
        if re.findall("(\d+)",line):
            atomz=re.findall("(\d+)",line)
            for z in atomz:
                pos1, pos2=0 , 0
                pos1, pos2=re.search(z,line).span()
                symbolset=[x for x in range(pos1, pos2)]
#                print(f"{pos1=}, {pos2=}, {symbolset=}")
                numlist.append(symbolset)
#                print(f"{numlist=}")

        symlist.append(symbolist)
        nulist.append(numlist)

print(symlist)
print(nulist)
print(rowlen)
print(linedict)
        
```
